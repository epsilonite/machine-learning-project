```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df_train = pd.read_csv('/content/drive/My Drive/machine-learning-project-local/data/df_train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('/content/drive/My Drive/machine-learning-project-local/data/df_test.csv')
df_test.head()
```

```python
# Separate features (X) and target (y) for training
X_train = df_train.drop(columns=['pathology'])  # Features
y_train = df_train['pathology']                 # Target variable

# Separate features (X) and target (y) for testing
X_test = df_test.drop(columns=['pathology'])    # Features
y_test = df_test['pathology']                   # Target variable
```

```python
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

# Align columns in train and test sets (ensure they have the same features)
X_train, X_test = X_train.align(X_test, axis=1, fill_value=0)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Initialize the KNN model
knn = KNeighborsClassifier(n_neighbors=5)

# Train the model
knn.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = knn.predict(X_test)
```

```python
# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Detailed Classification Report
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Parameter grid
param_grid = {'n_neighbors': range(1, 21)}

# Grid search with cross-validation
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Best number of neighbors
print("Best parameters:", grid_search.best_params_)
```
