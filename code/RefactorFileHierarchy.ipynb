```python
import os
import shutil
```

```python
# Root folder to perform operation
root = 'CBIS-DDSM/Mass-Test'
```

```python
# Walk through the directory tree from top to bottom
for (dirpath,folders,files) in os.walk(root):
    # At terminus of walk, or when files are found
    if len(files)>0:
        # Store parent folder that acts as key for csv/metadata
        folder = dirpath.split(os.sep)[1]
        print(folder)
        print('---------------------')
        # Construct the new filename using the parent folder
        filename = folder.lstrip('Mass-Test_')
        # Sort files by filesize in descending order
        files.sort(key=lambda x: os.path.getsize(os.path.join(dirpath, x)), reverse=True)
        for i,file in enumerate(files):
            # Set filename for fullsize image or fullsize mask
            new_filename = filename+'.dcm'
            # Set filename for cropped image if exists
            if i > 0: new_filename = filename+'_cropped.dcm'
            # Setup filepaths for move and rename
            old_file_path = os.path.join(dirpath, file)
            new_file_path = os.path.join(root, new_filename)
            # Move and rename the file to the root directory
            shutil.move(old_file_path, new_file_path)
            print(f"Moved and Renamed: {old_file_path} -> {new_file_path}")
```

```python
# Walk through the directory tree from bottom to top
for (dirpath,folders,files) in os.walk(root, topdown=False):
    for folder in folders:
        remove = os.path.join(dirpath, folder)
        # Check if the folder is empty
        if not os.listdir(remove):
            os.rmdir(remove)
            print(f"Removed empty folder: {remove}")
```
