```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dir = 'outputs/'
models = ['inceptionv3','resnet50','resnet50op','mobilenetv2op','mobilenetv2','nasnetmobile']
history = {}
accuracy = pd.DataFrame()
accuracy_val = pd.DataFrame()
loss = pd.DataFrame()
loss_val = pd.DataFrame()
```

```python
for model in models:
        history[model] = pd.read_csv(f"{dir}history-{model}.csv")
```

```python
for model in models:
    accuracy[f'{model}'] = history[model]['accuracy']
    accuracy_val[f'{model}'] = history[model]['val_accuracy']
    loss[f'{model}'] = history[model]['loss']
    loss_val[f'{model}'] = history[model]['val_loss']
```

```python
ax1 = accuracy.plot()
ax1.set_ylim(0.48, 1.02)
plt.savefig('images/accuracy.png')
```

```python
ax2 = loss.plot()
ax2.set_ylim(-0.02, 1.02)
plt.savefig('images/loss.png')
```

```python
ax3 = accuracy_val.plot()
ax3.set_ylim(0.48, 1.02)
plt.savefig('images/accuracy_val.png')
```

```python
ax4 = loss_val.plot()
ax4.set_ylim(0.28, 1.02)
plt.savefig('images/loss_val.png')
```
