```python
import pydicom
import numpy as np
import pandas as pd
```

```python
px_size=512
```

```python
# Load csv
df = pd.read_csv('data/df_test.csv')
# Set identifier as index
df.set_index('identifier',inplace=True)
```

```python
img_arrays = []
for id in df.index:
    # Get dcm for mask
    mask_path = f"CBIS-DDSM/Mass-Test/{id}.dcm"
    mask_array = pydicom.dcmread(mask_path).pixel_array
    # Get index of where mask is white
    mask = np.argwhere(mask_array == 255)
    # Get the first and last row and column indices where 255 occurs
    rowB = np.min(mask[:, 0])
    rowE = np.max(mask[:, 0])
    colB = np.min(mask[:, 1])
    colE = np.max(mask[:, 1])
    # Get start of square centered on mask of given dimension
    row_px = rowB - int((px_size-(rowE-rowB))/2)
    col_px = colB - int((px_size-(colE-colB))/2)
    # Get dcm for scan
    img_path = f"{mask_path.rsplit('_', 1)[0]}.dcm"
    # Get pixel array of region of interest
    img_arrays.append(pydicom.dcmread(img_path).pixel_array[row_px:row_px+px_size,col_px:col_px+px_size])
    print(id)
```
